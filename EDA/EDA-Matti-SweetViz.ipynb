Data Imports
importing unlabeled data as U1 importing labeled data as L1

```python
u1 = pd.read_csv("breast-cancer-wisconsin.data",header=None)
l1 = pd.read_csv("breast-cancer-wisconsin.data",header=None, names=['id','clump_thickness','uniformity_of_cell_size','uniformity_of_cell_shape','marginal_adhesion','single_epithelial_cell_size','bare_nuclei','bland_chromatin','normal_nucleoli','mitoses','class'])
print("u1")
print(u1.head())
print("l1")
print(l1.head())
```

```python
# split the dataset
x=l1.drop('class1',axis=1)
y=l1['class1']
x
```

```python
# Summary of the dataset
analysis = sv.analyze(l1)
analysis.show_html('EDA-Sweetviz.html', open_browser=False)
```

using Sklearn to split data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
#x_train.shape,x_test.shape,y_train.shape,y_test.shape
x_train
x_test
```

Comparison report

```python
compare_report = sv.compare([x_train, 'Train'], [x_test, 'Test'])
compare_report.show_html('EDA-Sweetviz-Compare.html', open_browser=False)
```
